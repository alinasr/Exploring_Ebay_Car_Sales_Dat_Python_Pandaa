## Finding common brands and their mileage

In this part we will find top brands and their distant traveled to see the relation


```python
unique_brands = autos["brand"].unique()
print(unique_brands)

autos["brand"].value_counts(normalize = True)
```

As you cans see the first brand is Volkswagen with 21% then BMW with 10%. Actually top 5 brands are German which shows the domination by brands from this country. Also we choose top 10 brands to aggregate by.

```python
top_10 = autos["brand"].value_counts(normalize = True)[:10].index

print(top_10)
```

```python
brands_mean_price = {}

for brand in top_10:
    mean_price = autos.loc[autos["brand"] == brand, 'price_dollar'].mean()
    brands_mean_price[brand] = mean_price
    
print(brands_mean_price)
```

As we can see the results the high mean price is or Audi brand and then Mercedes Benz then BMW.

We are going to fine mean mileage for top 10 brands as we did for mean price. The code will be like the mean price section


```python
brands_mean_milage = {}

for brand in top_10:
    mean_price = autos.loc[autos["brand"] == brand, 'odometer_km'].mean()
    brands_mean_milage[brand] = mean_price
    
print(brands_mean_milage)
```

In this part we are going to first convert both dictionaries into series object using series constructor then create data frame using first series by data frame constructor. Assign the other one to the data frame then print the data frame

```python
bmp_series = pd.Series(brands_mean_price)
bmm_series = pd.Series(brands_mean_milage)
df = pd.DataFrame(bmp_series, columns=['mean_price'])

df["mean_milage"] = bmm_series

print(df)
```

Very slightly the expensive cars have been drove more. But it is not very distinguishable.

## Extra cleaning

To clean the data set more we start first with translating German words to English.
Then we will convert string dates to numerical ones 
### German words to English

If we look at the data carefully we can realize in some columns there are German words and we can see the by the code blow by showing unique values for each that columns.


```python
print(autos['type_vehicle'].unique())
print(autos['gearbox'].unique())
print(autos['type_fuel'].unique())
print(autos['unrepaired_damage'].unique()) 
```

As we can see in above we shouel translate some words:

kleinwagen > small car

kombi > station wagon

andere > other

benzin > petrol

elektro > electrical

automatik > automatic

nein > no

ja > yes


in the code blow we replay the german words we their english counterparts

```python
autos['type_vehicle'] = autos['type_vehicle'].str.replace("kleinwagen", "small car")
autos['type_vehicle'] = autos['type_vehicle'].str.replace("kombi", "station wagon")

autos['gearbox'] = autos['gearbox'].str.replace("automatik", "automatic")

autos['type_fuel'] = autos['type_fuel'].str.replace("benzin", "petrol")
autos['type_fuel'] = autos['type_fuel'].str.replace("elektro", "other")
autos['type_fuel'] = autos['type_fuel'].str.replace("andere", "other")

autos['unrepaired_damage'] = autos['unrepaired_damage'].str.replace("nein", "no")
autos['unrepaired_damage'] = autos['unrepaired_damage'].str.replace("ja", "yes")

autos['model'] = autos['model'].str.replace("andere", "other")
```

### String dates to numerical values

In this section we will convert dates to numerical values for example

2016-05-25 to 20160525 

We first replace the '-' with empty character then we will remove the time after that we will convert the type to integer 
We did this changes to columns date_crawled, ad_created  and last_seen  as in code blow


```python
autos['date_crawled'] = autos['date_crawled'].str.replace("-", "")
autos['date_crawled'] = autos['date_crawled'].str[0:8].astype(int)

autos['ad_created'] = autos['ad_created'].str.replace("-", "")
autos['ad_created'] = autos['ad_created'].str[0:8].astype(int)

autos['last_seen'] = autos['last_seen'].str.replace("-", "")
autos['last_seen'] = autos['last_seen'].str[0:8].astype(int)
```

## finding most common brand/model combination 

In this step we first define empty dictionary
Then fin top 10 model
After that we find corresponding model brand and the model's frequency

For dictionary key we use brand + space + model name and for the value the size of brand_model sries witch is the number of that model. Then for better view we construct a data frame from brand_model_top.

Note that we remove the 'other' model from top models


```python
brand_model_top = {}


Top_10_model = autos["model"].value_counts(ascending = False)[:10]

#dropping other lable
Top_10_model = Top_10_model.drop(labels="other")

for m in Top_10_model.index:
    brand_model = autos.loc[autos["model"] == m, 'brand']
    brand = brand_model.iloc[0]
    brand_model_top[brand + ' ' + m] = brand_model.size
    



brand_model_series = pd.Series(brand_model_top)

df2 = pd.DataFrame(brand_model_series, columns=['counts'])
df2
```

As we can see the most common brand model is Volkswagen golf then BMW 3er and etc.
